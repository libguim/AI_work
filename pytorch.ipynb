```python
a=torch.FloatTensor([[1,2],[3,4]])
b=torch.FloatTensor([[2,2],[3,2]])
print(a,b)
```

```python
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a==b)
print(a!=b)
print(a**b)
print(a)
print(b)
print(a.mul_(b))
print(a)
```

```python
print(a.sum())
print(a.mean())
print(a.sum(dim=0))
print(a.sum(dim=1))
print(a.mean(dim=0))
print(a.mean(dim=1))
print(a.sum(dim=-1))
```
