## Class 로 구현

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
```

```python
torch.manual_seed(1)
x_data=torch.FloatTensor([[1],[2],[3],[4],[5]])
t_data=torch.FloatTensor([[3],[5],[7],[9],[11]])
```

```python
class LinearRegressionModel(nn.Module):
    def __init__(self): # 모델 생성
        super().__init__()
        self.linear=nn.Linear(1,1)

    def forward(self, x):
        return self.linear(x)
```

```python
model=LinearRegressionModel()
print(list(model.parameters()))
optimizer=optim.SGD(model.parameters(), lr=0.01)
```

```python
epochs=2001
for epoch in range(epochs):
    y=model(x_data)
    cost=F.mse_loss(y, t_data)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch%100==0:
        print(epoch, cost.item(), list(model.parameters()))
```

```python
import pandas as pd
df=pd.read_csv('test-data(score).csv')
data=np.array(df.values)
#print(data)
x_data=data[:,:3].reshape(-1, 3)
t_data=data[:,-1].reshape(-1, 1)
print(x_data)
print(t_data)
x_data3=torch.from_numpy(x_data).float()
t_data3=torch.from_numpy(t_data).float()
print(x_data3)
print(t_data3)
```

```python
class MultiLinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear=nn.Linear(3,1)

    def forward(self, x):
        return self.linear(x)
```

```python
model=MultiLinearRegressionModel()
optimizer=optim.SGD(model.parameters(), lr=1e-6)
```

```python
epochs=200001
for epoch in range(epochs):
    y=model(x_data3)
    cost=F.mse_loss(y, t_data3)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch%400==0:
        print(epoch, cost.item(), list(model.parameters()))
```
