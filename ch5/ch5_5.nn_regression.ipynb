```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
```

```python
torch.manual_seed(1)
x_data=torch.FloatTensor([[1],[2],[3],[4],[5]])
t_data=torch.FloatTensor([[3],[5],[7],[9],[11]])
```

### 입력 1개, 1변수, 단순회귀

```python
model=nn.Linear(1,1) # input_dim=1, output_dim=1  y=wx+b
print(list(model.parameters()))
```

```python
optimizer=optim.SGD(model.parameters(), lr=0.01)
epochs=2001

for epoch in range(epochs):
    h=model(x_data)
    cost=F.mse_loss(h,t_data)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    
    if epoch%100==0:
        print('epoch:',epoch, 'cost:',cost.item())
```

```python
x=torch.FloatTensor([3])
model(x)
```

### 입력 3개, 다변수, 다중회귀

```python
import pandas as pd
df=pd.read_csv('test-data(score).csv')
data=np.array(df.values)
#print(data)
x_data=data[:,:3].reshape(-1, 3)
t_data=data[:,-1].reshape(-1, 1)
print(x_data)
print(t_data)
x_data3=torch.from_numpy(x_data).float()
t_data3=torch.from_numpy(t_data).float()
print(x_data3)
print(t_data3)
```

```python
torch.manual_seed(1)
model=nn.Linear(3,1)
print(list(model.parameters()))
```

```python
optimizer=optim.SGD(model.parameters(), lr=1e-6)
epochs=20001
for epoch in range(epochs):
    y=model(x_data3)  #y=wx+b
    cost=F.mse_loss(y, t_data3)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch%400==0:
        print('epoch:',epoch, 'cost:',cost.item())
        print('Parameters:', list(model.parameters()))
```
