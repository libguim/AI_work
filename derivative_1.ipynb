```python
# 넘파이 적용확장 : 벡터(배열) 입력 지원

import numpy as np

def numerical_derivative(f,x):
    x=np.array(x, dtype=float)
    delta_x=1e-4

    return (f(x+delta_x) - f(x-delta_x)) / (2*delta_x)
```

```python
# 스칼라 입력

print(numerical_derivative(func1, 3.0))
```

```python
xs=[1.0,2.0,3.0,4.0]
print(numerical_derivative(func1, xs))
```

```python
# 다변수 함수 편미분 지원

def numerical_partial_derivative(f,x):
    """
    f : 다변수 함수, 입력은 1차원 넘파이 배열
    x : 미분을 구할 점(리스트나 배열)
    반환값 : 각 변수에 대한 편미분 값이 담긴 1차원 넘파이 배열
    """
    x=np.array(x, dtype=float)
    delta_x=1e-4
    grad=np.zeros_like(x)

    for i in range(len(x)):
        x1=np.copy(x)
        x2=np.copy(x)

        x1[i] += delta_x
        x2[i] -= delta_x

        grad[i] = (f(x1) - f(x2)) / (2* delta_x)

    return grad
```

```python
# 2변수 함수
def func(v):
    x, y = v
    return x**2 + y**3

point=[3.0,2.0]
print(numerical_partial_derivative(func, point))
```

```python
# 3변수 함수
def func3(v):
    x,y,z=v
    return x*y + np.sin(z)

print(numerical_partial_derivative(func3, [1.0,2.0, np.pi/4]))
```

```python
# 다변수 함수 + 여러점 동시계산

def batch_numerical_partial_derivative(f, points):
    """
    f : 다변수 함수, 입력은 1차원 넘파이 배열
    points : 여러 점의 좌표 (2D 배열 - like), shape : (n_points, n_variables)
    반환값 : 각 점마다의 그래디언트 벡터 (shape : (n_points, n_variables))
    """
    points=np.atleast_2d(points).astype(float)
    n_points, n_vars = points.shape
    delta_x=1e-4
    grads=np.zeros((n_points, n_vars))

    for p_idx, point in enumerate(points):
        grad=np.zeros(n_vars)
        for i in range(n_vars):
            x1=np.copy(point)
            x2=np.copy(point)
            x1[i] += delta_x
            x2[i] -= delta_x
            grad[i] = (f(x1) - f(x2)) / (2*delta_x)
        grads[p_idx] = grad

    return grads
```

```python
# 2변수 함수
def func(v):
    x, y = v
    return x**2 + y**3

points=[
    [3.0,2.0],
    [1.0,-1.0],
    [0.0,0.0],
]

print(batch_numerical_partial_derivative(func, points))
```
